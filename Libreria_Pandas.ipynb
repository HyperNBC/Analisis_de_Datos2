```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names) # Crear primero el dataframe
iris_df['target'] = iris.target #Agregar en la columna de especies con la informacion del campo target
iris_df
```

```python
iris_df.columns
```

```python
#Renombramos la columna target para species
iris_df.rename({"target": "species"}, inplace=True, axis=1)
iris_df
```

```python
iris_df["species"] = iris_df["species"].astype("category")
iris_df.info()
```

```python
#Asigno valores nulos a la columna de especies
#Perc=0.25 significa que voy a asignar valores nulos al 25% de los datos
iris_amp = mf.ampute_data(iris_df, perc=0.25, random_state=1991) #Realizo una amputación de datos aplicando NaN
iris_amp
```

```python
#Crear Kernel
kds = mf.ImputationKernel(
    iris_amp,                #Nombre de dataframe para entrenar modelo
    random_state=1991        #Valores aleatorios para entrenar modelo
)

#Correr el algoritmo MICE durante 2 interacciones
kds.mice(2)

#Retorna el dataset completo
iris_completo = kds.complete_data()
iris_completo
```
